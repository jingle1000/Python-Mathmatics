### Newtons Method
------------------------------


```python
import numpy as np
from scipy.misc import derivative
import matplotlib.pyplot as plt
```

```python
x_n = 2 #starting guess
x = np.linspace(0.2, 2.2, 500) #not starting at zero because natural log is undefined at zero

f = lambda x: np.log(x)

fx_next = lambda x_n, funct: x_n - funct(x_n) / derivative(funct, x_n, dx=0.01)
    
```

```python
for i in range(1, 6):
    print(fx_next(x_n, f))
```
