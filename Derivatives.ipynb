```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sympy as sy
import math
```

```python
'''Plotting the derivitive of sin(x)'''

N = 10000 #num function evaluations
x = np.linspace(0, 4*math.pi, N)

h = (4*math.pi) / N * 1.0 #make sure that python knows h will be a float

def f(x):
    val = math.sin(x)
    return val

def fPrime(x):
    return (f(x+h) - f(x)) / h

y2 = [] #the y values for the tangent line
for i in x:
    derivitive = fPrime(i)
    y2.append(derivitive)

y1 = []
for i in x:
    y1.append(f(i))

plt.plot(x, y1)
plt.plot(x, y2)
```

```python
N = 10000
x = np.linspace(-10, 10, N)

def f(x):
    return 3*(x ** 4) - 5*(x**3)+(6*x)+8

def fPrime(x):
    return 3*(4*(x**3)) - 5*(3*(x**2))+6

def tangentLine(m, x, b):
    return m * x + b

def solveB(x, y, m):
    return y - (m*x)

atX = 7 #the x value of the point we are trying to find
atY = f(atX) #get the y value of the point on the equation

m = fPrime(atX) #get the slope
b = solveB(atX, atY, m) #get b

y1 = [] #the regular y values
for i in x:
    y1.append(f(i))
    
y2 = [] #yvalues for the tangent line
for i in x:
    y2.append(tangentLine(m, i, b))

plt.plot(x, y1)
plt.plot(x, y2)
```
