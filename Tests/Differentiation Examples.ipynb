```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import *
import cmath
import sys
```

```python
x = [0, 1, 2, 3]
fx = [4, 0, -2, -1]
fprimex = [2, 3, 5, -2]
gx = [3, 2, 0, -3]
gprimex = [-3, 1, -1, 4]

z = np.polyfit(x, fx, 3)
p = np.poly1d(z)
x1 = np.linspace(-4, 4)
y1 = p(x1)
```

```python
plt.plot(x, fx, x1, y1)
```

```python
pprime = p.deriv()
primey = pprime(x1)
```

```python
plt.plot(x1, y1, x1, primey)
```
