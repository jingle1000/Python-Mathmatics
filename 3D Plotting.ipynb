```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.tri import Triangulation
import matplotlib.pyplot as plt
from scipy import interpolate
from scipy.integrate import ode
import numpy as np

import plotly.graph_objs as go
#from IPython import get_ipython
#get_ipython().run_line_magic('matplotlib', 'qt')
```

```python
'''start in 2d by drawing a basic circle'''

theta = np.linspace(0, 2*np.pi, 40)
#radius
r = np.sqrt(10)
#x and y coordinates
x = r*np.sin(theta)
y = r*np.cos(theta)

#plot the circle
plt.gca().set_aspect('equal')
plt.plot(x, y)
plt.show()#a basic circle
theta = np.linspace(0, 2*np.pi, 40)
```

```python
#mesh the data together
Xc, Yc = np.meshgrid(x, theta)
Yc, Zc = np.meshgrid(y, theta)

#plot the "cylinder" aka -> series of vertical lines in a circular pattern
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(Xc, Yc, Zc,cmap='viridis', edgecolor='none')
plt.show()
# ax.scatter(x, y, z)
```

```python
fig = go.Figure(data=[go.Surface(Xc, Yc, Zc)])
```
